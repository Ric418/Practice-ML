```python
import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
import random
import torch
import torch.nn as nn
```

```python
x = np.arange(-8, 8, 0.1) #

#　ランプ関数
y = x * (x > 0)
fig1, DGx = plt.subplots()
DGx.set_xlabel('input')
DGx.set_ylabel('output')
# グラフの設定
plt.plot(x, y, lw=5) # プロット
plt.xlim(-8, 8)  # x軸の範囲
plt.ylim(-2, 8) # y軸の範囲
plt.grid() # グリッド描画
plt.show() # グラフを出力

fig1.savefig('ReLu.png')
```

```python
x=torch.linspace(-8,8,200,dtype=torch.float)
fig1, DGx = plt.subplots()
DGx.set_xlabel('input')
DGx.set_ylabel('output')

m=nn.LeakyReLU(0.1)
y=m(x)

plt.plot(x.numpy(), y.numpy(), lw=5) # プロット
plt.xlim(-8, 8)  # x軸の範囲
plt.ylim(-2, 8) # y軸の範囲
plt.grid() # グリッド描画
plt.show() # グラフを出力

fig1.savefig('LeakyReLu.png')
```

```python
x=torch.linspace(-8,8,200,dtype=torch.float)
fig1, DGx = plt.subplots()
DGx.set_xlabel('input')
DGx.set_ylabel('output')

m=nn.Tanh()
y=m(x)

plt.plot(x.numpy(), y.numpy(), lw=5) # プロット
plt.xlim(-8, 8)  # x軸の範囲
plt.ylim(-1.1, 1.1) # y軸の範囲
plt.grid() # グリッド描画
plt.show() # グラフを出力

fig1.savefig('Tanh.png')
```
