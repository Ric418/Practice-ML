```python
import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
import random
import torch
```

```python
n_epoch = 50
test_num=np.random.random(540)
```

```python
test_num2=np.random.random(540)
```

```python
D_loss = []
G_loss = []  

print(test_num)
print(test_num2)
```

```python
test_torch=torch.from_numpy(test_num)
test_torch2=torch.from_numpy(test_num2)
test_torch.cuda()
print(test_torch)
print(test_torch2)
print(range(test_torch.shape[0]))
print(range(test_torch2.shape[0]))
```

```python
for itr in range(test_torch.shape[0]):
    D_loss.append(test_torch.numpy()[itr])
    G_loss.append(test_torch2.numpy()[itr])
```

```python
x_num = list(map(lambda x: x*100, np.arange(0,540)))

print(x_num)
matplotlib.style.use('ggplot')
plt.figure(figsize=(2000, 2000), dpi=10)
fig1, DGx = plt.subplots()
DGx.set_xlabel('epoch')
DGx.set_ylabel('Loss')
DGx.set_xlim(0, 54000)

plt.xticks(np.arange(0, 54000 + 1, 10000))
DGx.plot(x_num,D_loss,label='Discriminator Loss')
DGx.plot(x_num,G_loss,label='Generater Loss')
DGx.legend(loc = 'best')

# fig1.savefig('Losses.png')
# fig2.savefig('Dis_Var.png')

```
